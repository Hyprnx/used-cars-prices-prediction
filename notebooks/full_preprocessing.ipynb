```python
import numpy as np
import pandas as pd
import json
import pandas_profiling as pp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '/content/chotot_used_cars.csv'
path_1 = '/content/bonbanh_used_cars.json'
path_2 = '/content/otocom_used_car.json'
path_3 = '/content/sanxehot_detail.json'
f = open(path_1, 'r')
f1 = open(path_2, 'r')
f2 = open(path_3, 'r')
data1 = json.load(f)
data2 = json.load(f1)
data3 = json.load(f2)
chotot = pd.read_csv(path)
bonbanh = pd.DataFrame.from_records(data1)
otocom = pd.DataFrame.from_records(data2)
sanxehot = pd.DataFrame.from_records(data3)
```

##Standardized data sets

```python
chotot['type'] = chotot['type'].map({'Sedan':'sedan', 'Hatchback' : 'hatchback', 'Van' : 'van', 'SUV / Cross over': 'suv', 'Minivan (MPV)': 'mpv', 'Pick-up (bán tải)' : 'pick-up', 'Kiểu dáng khác' : np.nan, 'Coupe (2 cửa)':'coupe', 'Mui trần' : 'convertible', '' : np.nan})
```

```python
otocom.transmission = otocom.transmission.apply(lambda x: x.strip())
otocom = otocom[otocom.transmission.isin(['automatic', 'manual', 'semi-automatic'])]
otocom.transmission.unique()
```

```python
chotot = chotot[(chotot.transmission != '5')]
```

```python
otocom['type'] = otocom.type.apply(lambda x: x.strip())
otocom.replace('pick-up truck', 'pickup', inplace=True)
otocom.type.unique()
```

```python
otocom = otocom[otocom.type != 'sport car']
```

```python
chotot.replace('Động cơ hybrid', 'hybrid', inplace = True)
chotot.fuel.unique()
```

```python
bonbanh.drop(['internal_color', 'external_color', 'name'], axis = 1, inplace=True)
bonbanh_source = bonbanh.source_url
bonbanh.drop(['source_url'], axis = 1, inplace=True)
```

```python
otocom.rename(columns={'Engine displacement': 'engine_capacity', 'name': 'model', 'model': 'type'}, inplace=True)
otocom_source = otocom.source_url
otocom_city = otocom.city
otocom.drop(['source_url', 'city'], axis = 1, inplace = True)
```

```python
sanxehot = sanxehot[sanxehot['name'].notna()]
sanxehot = sanxehot[sanxehot['year'].notna()]
sanxehot_source_url = sanxehot.source_url
sanxehot.name = sanxehot.name.str.replace('-',' ')
sanxehot['model'] = [a.replace(b, '').strip() for a, b in zip(sanxehot['name'], sanxehot['brand'])]
sanxehot.model = sanxehot.model.str.replace('Morning / Picanto', 'Morning')
sanxehot.model.unique()
sanxehot
```

```python
sanxehot_source_url = sanxehot.source_url
sanxehot.drop(['name', 'source_url', 'external_color', 'wheel_drive'], axis=1, inplace=True)
```

```python
chotot.drop('url', axis = 1, inplace=True)
chotot.rename({'fuel': 'fuels', 'date': 'year', 'prices':'price'}, axis = 1, inplace = True)
chotot.head()
```

```python
bonbanh.sort_index(axis=1, inplace=True)
otocom.sort_index(axis=1, inplace=True)
sanxehot.sort_index(axis=1, inplace=True)
chotot.sort_index(axis=1, inplace=True)
```

```python
bonbanh.drop(['wheel_drive'], axis = 1 , inplace = True) # remove this when all other dataset filled with this missing info
```

##Combined dataset

```python
cars = pd.concat([bonbanh,  otocom, sanxehot, chotot])
```

```python
cars.dropna(subset=['brand'],inplace=True)
cars.brand = cars.brand.apply(lambda x: x.lower().strip())
```

```python
cars.model.unique()
cars.model = cars.model.apply(lambda x: x.lower().strip())
cars.model.replace({'khác':np.nan, 'Khác':np.nan, 'dòng':np.nan},inplace=True)
# cars.brand.replace({'hãng khác':np.nan},inplace=True)
cars.model.unique()
```

```python
cars.reset_index(drop=True, inplace=True)
```

```python
khac = cars[cars.brand == 'hãng khác'].index
cars.drop(index=khac, inplace=True)
```

```python
cars.replace(['mercedes benz', 'mercedes-benz'], 'mercedes', inplace=True)
```

```python
#process BMW & Mercedes
cars.replace('series 3', '3 series', inplace = True)
cars.replace('series 5', '5 series', inplace = True)
bmw_index = cars[cars['brand'] == 'bmw'].index
merc_index = cars[cars['brand'] == 'mercedes'].index
# bmw.model = bmw.model.str.split(expand=True)[0]
# bmw.model.value_counts()[:15]
cars.model.loc[([*bmw_index, *merc_index])] = cars.loc[([*bmw_index, *merc_index])].model.str.split(expand=True)[0]
```

```python
#process Honda & Mazda

cars.replace(['cr', 'cr v', 'cr-v'], 'crv', inplace = True)
cars.replace(['cx-5', 'cx 5'], 'cx5', inplace = True)
cars.replace(['bt-50', 'bt 50'], 'bt50', inplace = True)
cars.replace(['cx-30', 'cx 30'], 'cx30', inplace = True)
cars.replace(['cx-8', 'cx 8'], 'cx8', inplace = True)
cars.replace(['cx-9', 'cx 9'], 'cx9', inplace = True)
cars.replace('hr-v', 'hrv', inplace = True)
```

```python
cars.replace('santa', 'santafe', inplace = True)
```

```python
cars = cars[(cars['model'] != 'cx')]
cars
```

```python
#process Vinfast

lux_a = cars[(cars['brand'] == 'vinfast') & (cars['model'] == 'lux') & (cars['type'] == 'sedan')].index
lux_sa = cars[(cars['brand'] == 'vinfast') & (cars['model'] == 'lux') & (cars['type'] == 'suv')].index
cars.model.loc[lux_a] = 'lux a 2.0'
cars.model.loc[lux_sa] = 'lux sa 2.0'
```

```python
#process Toyota

drop_corolla = [*cars[(cars['brand'] == 'toyota') & (cars['model'] == 'corolla') & (cars['type'] == 'mpv')].index, *cars[(cars['brand'] == 'toyota') & (cars['model'] == 'corolla') & (cars['type'] == 'hatchback')].index]
cars.drop(index=drop_corolla, inplace = True)
cars.replace('altis', 'corolla altis', inplace = True)
```

```python
top15 = cars.brand.value_counts()[:15].index
cars1 = cars.loc[cars['brand'].isin(top15)]
cars1.shape
```

```python
cars[(cars['brand'] == 'toyota') & (cars['model'] == 'corolla')].type.unique()
```

```python
corolla_suv = cars[(cars['brand'] == 'toyota') & (cars['model'] == 'corolla') & (cars['type'] == 'suv')].index
corolla_sedan = cars[(cars['brand'] == 'toyota') & (cars['model'] == 'corolla') & (cars['type'] == 'sedan')].index
cars.model.loc[corolla_suv] = 'corolla cross'
cars.model.loc[corolla_sedan] = 'corolla altis'
```

```python
cars_process = cars.copy()
cars_process.model = cars['brand'] + ' ' + cars['model']
cars_process
```

```python
#keep only top 15 most appear model each brand
model_keep = []

def top_appear(brand):
  _keep_index = cars_process[cars_process['brand']==brand].model.value_counts()[:15].index
  _keep_values = cars_process[cars_process['brand']==brand].model.value_counts()[:15]

  for i, j in zip(_keep_index, _keep_values):
    if j > 27:
      model_keep.append(i)
      # print('{} : {}'.format(i,j))

model_keep = [*model_keep, *cars_process[cars_process['brand'] == 'vinfast'].model.unique()]

for _ in top15:
  top_appear(_)

cars_process = cars_process.loc[cars_process['model'].isin(model_keep)]
cars_process
```

```python
cars_process = cars_process[cars_process.km_driven > 1000]
cars_process.shape
```

```python
cars_process.reset_index(drop=True, inplace=True)
```

```python
from scipy import stats

q_low_price = cars_process["price"].quantile(0.02)
q_hi_price  = cars_process["price"].quantile(0.97)

cars_process = cars_process[(cars_process["price"] < q_hi_price) & (cars_process["price"] > q_low_price)]
cars_process.shape
```

```python
cars_process = cars_process[cars_process['year'] > 2007]
```

```python
cars_process.dropna(inplace=True)
```

##Encoding categorical

```python
transmission = cars_process['transmission']
transmission = pd.get_dummies(transmission,drop_first=True)
transmission.head()
```

```python
fuels = cars_process['fuels']
fuels = pd.get_dummies(fuels,drop_first=True)
fuels.head()
```

```python
cars_process.replace('pick-up', 'pickup', inplace=True)
```

```python
type1 = cars_process['type']
type1 = pd.get_dummies(type1,drop_first=True)
type1.head()
```

```python
origin = cars_process['origin']
origin = pd.get_dummies(origin,drop_first=True)
origin.head()
```

```python
final_cars = pd.concat([cars_process, transmission, fuels, origin, type1],axis = 1)
final_cars.shape
```

```python
#Save data to csv
final_cars.to_csv('final_cars.csv', index = False)
```
